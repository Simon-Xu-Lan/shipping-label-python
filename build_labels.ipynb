```python
cols = ['Shipping Country','WarehoUS - Transgroupe','CONFIG', 'Carton Type', 'ORDER #', 'Carton of Total',
       'SKU-D9-56EXPANSION', 'SKU-D9-ALLY-PACK', 'SKU-D9-BASE-GAME',
       'SKU-D9-EXTRA-MINIS', 'SKU-D9-PUDDI-CAN', 'SKU-D9-SLEEVES',
       'SKU-D9-TOKENS', 'Total items in Carton']
```

```python
def create_label(row, cols):
    pdf = FPDF(orientation='p', unit='mm', format=(160, 180))
    pdf.add_page()
    pdf.set_font("Arial", size=14)
    
    for col in cols:
        if col.startswith("SKU") and math.isnan(row[col]):
            continue
        top = pdf.y
        offset = pdf.x + 80
        pdf.multi_cell(80, 10, col, 0, 0)
        pdf.y = top
        pdf.x = offset
        if col.startswith("SKU"): 
            pdf.multi_cell(100, 10, str(int(row[col])))
        else:
            pdf.multi_cell(100, 10, str(row[col]))
    country_code = row["WarehoUS - Transgroupe"][:2]
    filename = f"./labels/{country_code}-{row['CONFIG']}-{row['ORDER #']}-{row['Carton of Total']}.pdf"
    pdf.output(filename, "F")
```

```python
for index, row in df.iterrows():
    create_label(row, cols)
```
