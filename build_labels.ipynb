```python
from fpdf import FPDF
import pandas as pd
import numpy as np
import math
```

```python
df = pd.read_excel('D9 BOARDGAME FACTORY PACKING LIST_FINAL.xlsx', sheet_name='data', header=1)
df
```

```python
df.head()
```

```python
cols = df.columns
```

```python
def create_label(row, cols):
    pdf = FPDF(orientation='p', unit='mm', format=(150, 160))
    pdf.add_page()
    pdf.set_font("Arial", size=14)
    
    for col in cols:
        if col.startswith("SKU") and math.isnan(row[col]):
            continue
        top = pdf.y
        offset = pdf.x + 80
        pdf.multi_cell(80, 10, col, 0, 0)
        pdf.y = top
        pdf.x = offset
        if col.startswith("SKU"): 
            pdf.multi_cell(100, 10, str(int(row[col])))
        else:
            pdf.multi_cell(100, 10, str(row[col]))
    filename = f"./labels/label-{row['CONFIG']}-{row['ORDER #']}.pdf"
    pdf.output(filename, "F")
```

```python
for index, row in df.iterrows():
    create_label(row, cols)
```
